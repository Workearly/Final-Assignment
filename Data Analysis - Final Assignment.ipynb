```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading the data
data = pd.read_csv('liquor_sales_2016_2019.csv')
```

### most popular item sold based on zip code

```python
popularity_zip_code = pd.DataFrame(data.groupby(by=['zip_code','item_description']).sum()['bottles_sold']).reset_index()
```

```python
# top 5 zipcodes in popularity
popularity_zip_code.head()
```

```python
popularity_zip_code.sort_values(by=['zip_code', 'bottles_sold'],  ascending = [True, False])
```

```python
sns.scatterplot(data=popularity_zip_code.reset_index(), x="zip_code", y="bottles_sold", hue='item_description', legend=False)
plt.ylabel('Bottles Sold')
plt.title('Bottles Sold')
plt.show()
```

```python
popularity_zip_code.to_csv('popularity_zip_code.csv')
```

### percentage of sales per store

```python
sales_per_store = pd.DataFrame(data.groupby(by=['store_number','store_name']).sum()['sale_dollars']).reset_index()
```

```python
sales_per_store.head()
```

```python
sales_per_store['sale_dollars_percentage'] = sales_per_store['sale_dollars'].apply(lambda x: (x/data['sale_dollars'].sum())*100)
```

```python
sales_per_store.sort_values(by='sale_dollars_percentage', ascending=False)
```

```python
sales_per_store['store_number'] = sales_per_store['store_number'].astype(str)
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=sales_per_store.sort_values(by='sale_dollars_percentage', ascending=False), x="sale_dollars_percentage", y="store_number",palette="pastel")
plt.ylabel('Store Number')
plt.xlabel('Percentage of Total Sales')
plt.show()
```

```python
sales_per_store.to_csv('sales_per_store.csv')
```
